```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
import xgboost as xgb
import lightgbm as lgb
from imblearn.over_sampling import SMOTE
import pickle
```

```python
df = pd.read_csv(r"C:\Users\azaru\Azar documents\Data science - GUVI\Final project\MDT41 [DS-WE-T-B2]-20250717T133437Z-1-001\MDT41 [DS-WE-T-B2]\cleaned_dataset.csv")
```

```python
lab_enc = LabelEncoder()
df['y'] = lab_enc.fit_transform(df['y']) 

X = df.drop('y', axis=1)
X = pd.get_dummies(X, drop_first=True).astype(int)
y = df['y']

x = X.reset_index(drop=True)
y = y.reset_index(drop=True)


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)

smote = SMOTE(random_state=42)
x_train_smote,y_train_smote = smote.fit_resample(x_train,y_train)

models = [
    DecisionTreeClassifier(random_state=42),
    RandomForestClassifier(random_state=42),
    xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    lgb.LGBMClassifier(random_state=42)]

for model in models:
    model.fit(x_train_smote,y_train_smote)

    train_predection = model.predict(x_train)
    test_predection = model.predict(x_test)

    print(f"*********{type(model).__name__}**********")
    print(f"accuracy Train : {accuracy_score(y_train,train_predection)}")
    print(f"accuracy Test : {accuracy_score(y_test,test_predection)}\n")

    print(f"Precition Train : {precision_score(y_train,train_predection)}")
    print(f"Precition Test : {precision_score(y_test,test_predection)}\n") 

    print(f"recall Train : {recall_score(y_train,train_predection)}")
    print(f"recall Test : {recall_score(y_test,test_predection)}\n") 

    print(f"f1 Train : {f1_score(y_train,train_predection)}")
    print(f"f1 Test : {f1_score(y_test,test_predection)}\n") 
```

```python
lab_enc = LabelEncoder()
df['y'] = lab_enc.fit_transform(df['y']) 

X = df.drop('y', axis=1)
X = pd.get_dummies(X, drop_first=True).astype(int)
y = df['y']

x = X.reset_index(drop=True)
y = y.reset_index(drop=True)


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)

smote = SMOTE(random_state=42)
x_train_smote,y_train_smote = smote.fit_resample(x_train,y_train)

model = lgb.LGBMClassifier(random_state=42)

model.fit(x_train_smote,y_train_smote)

train_predection = model.predict(x_train)
test_predection = model.predict(x_test)

print(f"*********{type(model).__name__}**********")
print(f"accuracy Train : {accuracy_score(y_train,train_predection)}")
print(f"accuracy Test : {accuracy_score(y_test,test_predection)}\n")

print(f"Precition Train : {precision_score(y_train,train_predection)}")
print(f"Precition Test : {precision_score(y_test,test_predection)}\n") 

print(f"recall Train : {recall_score(y_train,train_predection)}")
print(f"recall Test : {recall_score(y_test,test_predection)}\n") 

print(f"f1 Train : {f1_score(y_train,train_predection)}")
print(f"f1 Test : {f1_score(y_test,test_predection)}\n") 

with open('Term deposit subscription Prediction.pkl','wb') as ft:
    pickle.dump(model,ft)

```

```python
with open('Term deposit subscription Prediction.pkl', 'rb') as ft:
    model = pickle.load(ft)

input_dict = {
    'age': 58,
    'job': 'management',
    'marital': 'married',
    'education': 'tertiary',
    'default': 'no',
    'balance': 2143,
    'housing': 'yes',
    'loan': 'no',
    'contact_day': 5,
    'contact_month': 'may',
    'duration': 261,
    'campaign': 1,
    'pdays': -1,
    'previous': 0
}

input_df = pd.DataFrame([input_dict])

binary_cols = ['default', 'housing', 'loan']
lab_enc = LabelEncoder()
for col in binary_cols:
    input_df[col] = lab_enc.fit(['no', 'yes']).transform(input_df[col])

input_df = pd.get_dummies(input_df, drop_first=True)

expected_cols = model.feature_names_in_
for col in expected_cols:
    if col not in input_df.columns:
        input_df[col] = 0
input_df = input_df[expected_cols] 

result = model.predict(input_df)

print(f"Is user Subscribed : {'Yes' if result[0] == 1 else 'No'}")
```
